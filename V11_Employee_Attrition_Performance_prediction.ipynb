## 1. Problem Statement & Project Overview

### 1.1 Business Problem

Employee attrition costs companies significant resources in recruitment, training, and lost productivity. Studies show that replacing an employee costs approximately **1.5 to 2 times their annual salary**. For organizations, the ability to predict which employees are likely to leave enables proactive retention strategies and substantial cost savings.

### 1.2 Machine Learning Task
 
* **Type of Learning:** Supervised Learning  
* **Task Type:** Binary Classification  
* **Target Variable:** Attrition (Yes/No)  
* **Primary Goal:** Predict employee attrition with high recall to identify at-risk employees before they leave

### 1.3 Why This Matters

* **Financial Impact:** The average cost to replace an employee ranges from $15,000 to $50,000+ depending on role and seniority
* **Productivity Loss:** It takes 6-12 months for a new employee to reach full productivity
* **Team Morale:** High turnover negatively impacts remaining employees and organizational culture
* **Early Detection:** Predictive models enable HR departments to implement targeted interventions 3-6 months before potential departures

### 1.4 Project Goals

1. **Build and compare multiple classification models** to identify the best performer.
2. **Identify key features** that predict employee attrition through feature importance analysis.
3. **Achieve high recall (>80%)** to minimize false negatives (missing at-risk employees).
4. **Provide actionable insights** for HR decision-making and retention strategies.
5. **Handle class imbalance** effectively using techniques like SMOTE.

### 5.5 Check outliers

```python
cols_exclude = ['Education', 'JobLevel', 'StandardHours', 'EmployeeNumber', 'EmployeeCount', 'StockOptionLevel', 'JobSatisfaction', 'EnvironmentSatisfaction', 
                'RelationshipSatisfaction', 'JobInvolvement', 'WorkLifeBalance', 'PerformanceRating']

features_to_check_outlier = [cols for cols in numerical_feature if cols not in cols_exclude]
print(features_to_check_outlier)
```

```python
def outlier_detection(df, features, mode='none'):
    """
    mode options:
    - 'none': Just print stats (default).
    - 'remove': Drop rows with outliers.
    - 'cap': Replace outliers with lower/upper bounds.
    """
    
    df_processed = df.copy()        # We take a copy of the data so that we can work on it without corrupting the original data.
    rows_with_outliers = set()      # We prepare a "Set" to collect the row numbers (Indices) that contain outliers, in case we want to delete them later.
    outlier_summary = []            # We prepare an empty list to store the statistics for each column (how many outliers, their percentage, etc.) in order to present them as a report at the end.

    print(f'Starting Outlier Detection (Mode: {mode})...\n')

    for col in features:
        
        # 1. Calculating IQR Bounds
        Q1 = df_processed[col].quantile(0.25)
        Q3 = df_processed[col].quantile(0.75)
        IQR = Q3 - Q1

        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # 2. Discovering Outliers (for statistics only)
        # This line performs a filter. It tells the DataFrame: "Get me all the rows whose value in this column is less than the minimum or greater than the maximum."
        # If the code finds outlier values, it takes some information about them (their number, percentage, and the limits we calculated) and puts them in the outlier_summary list.
        outliers = df_processed[(df_processed[col] < lower_bound) | (df_processed[col] > upper_bound)]
        outlier_count = len(outliers)
        outlier_pct = (outlier_count / len(df_processed)) * 100 

        if outlier_count > 0:
            
            outlier_summary.append({
            'Feature': col,
            'Outlier_Count': outlier_count,
            'Percentage': f"{outlier_pct:.2f}%",
            'Q1': Q1,
            'Q3': Q3,
            'IQR': round(IQR, 2),
            'Lower_Bound': round(lower_bound, 2),
            'Upper_Bound': round(upper_bound, 2)
            })
            
            # # 3. Grouping the indices if we use the remove mode
            # if mode == 'remove':
            rows_with_outliers.update(outliers.index)
                
            # 4. Applying Capping (if the mode is cap)
            if mode == 'cap':
                # Any value less than lower will be lower, and any value greater than upper will be upper
                df_processed[col] = df_processed[col].clip(lower=lower_bound, upper=upper_bound)
            
            
    
    
    if len(outlier_summary) > 0:
        outlier_df = pd.DataFrame(outlier_summary).sort_values('Outlier_Count', ascending=False)
        print("\nOutliers Detected (sorted by count):\n")
        print(outlier_df.to_string(index=False))

        total_outliers = outlier_df['Outlier_Count'].sum()
        unique_rows_affected = len(rows_with_outliers)
        print(f"\nTotal outliers detected: {total_outliers} across {len(outlier_df)} features")
        print(f"Total unique rows affected: {unique_rows_affected}")
        print(f'Percentage of outliers in the data: {(unique_rows_affected / len(df_processed)) * 100:.2f} %')
    
    else:
        print('\nNo outliers detected in the provided features')
    
    
    
    # 4. Apply the delete (if the mod is remove)
    if mode == 'remove':
        
        if len(rows_with_outliers) > 0:
            
            print(f'-> Removing {len(rows_with_outliers)} rows with outliers...')
            df_processed = df_processed.drop(index=list(rows_with_outliers))
        
        else:
            print('-> No outliers to remove')
    
    
    elif mode == 'cap':
        print('\n-> Outliers have been capped to upper/lower bounds.')
    
    return df_processed




outlier_detection(df_clean, features_to_check_outlier)
```

### 7.2 Data Preprocessing Pipeline (`build_preprocessor`)

Before feeding the data into our machine learning models, we must clean and format it appropriately. Machine learning algorithms require numerical input and cannot handle missing values natively. 

This function dynamically builds a **scikit-learn `ColumnTransformer`** to process features based on their data types:

##### 🔢 1. Numeric Feature Pipeline (`num_cols`)
Applied to all integer and float columns (e.g., `Age`, `MonthlyIncome`):

* **Scaling (`StandardScaler`):** Standardizes features by removing the mean and scaling to unit variance. This ensures that features with larger numerical ranges do not overpower features with smaller ranges during model training.

##### 🔠 2. Categorical Feature Pipeline (`cat_cols`)
Applied to all object and category columns (e.g., `Department`, `BusinessTravel`):
* **Encoding (`OneHotEncoder`):** Converts nominal text data into a binary format (0s and 1s). This is crucial to prevent the model from falsely assuming any mathematical or ranked relationship between independent categories.

**Returns:**
* `preprocess`: The compiled pipeline object ready to be integrated into our modeling workflow.
* `num_cols`: A generated list of the numeric feature names.
* `cat_cols`: A generated list of the categorical feature names.

### 7.3 Modeling & Evaluation

```python
def build_preprocessing(x):
    
    """This function dynamically builds a **scikit-learn `ColumnTransformer`** to process features based on their data types"""
    
    cat_cols = [col for col in x.columns if x[col].dtype == 'object' or str(x[col].dtype).startswith('category')]
    num_cols = [col for col in x.columns if col not in cat_cols]
    
    
    numerical_transform = Pipeline(steps=[
        ('scaler', RobustScaler())
    ])
    
    category_transform = Pipeline(steps=[
        ('encoding_onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
    ])
    
    
    preprocessor = ColumnTransformer(
        transformers=[
            
            ('num', numerical_transform, num_cols),
            ('cat', category_transform, cat_cols)
        ],
        remainder='drop'
    )
    
    
    return preprocessor, num_cols, cat_cols
```

```python
def experiment(x, y):
    
    """Train/tune several models and return a results table plus best pipeline."""
    
    preprocess, num_cols, cat_cols = build_preprocessing(x=x)
    
    models_and_grids = {
        
        "LogReg" : (
            # More attempts because saga and L1 need time to reach the best result (to avoid ConvergenceWarning)
            LogisticRegression(max_iter=1000, class_weight='balanced', random_state=42),
            {
                # 'C' controls the strength of regularization
                # Small values ​​(e.g., 0.01 and 0.1) indicate stronger regularization (strongly prevents overfitting)
                # Large values ​​(e.g., 1 and 10) indicate weaker regularization
                'model__C': [0.1, 1, 10],

                # 'l1_ratio' specifies the type of regularization instead of a penalty:
                # 0.0 = Full L2 application (Ridge)
                # 1.0 = Full L1 application (Lasso - removes unnecessary features)
                # 0.5 = 50% L1 and 50% L2 mix (called ElasticNet)
                'model__l1_ratio' : [0.0, 0.5, 1.0],
                'model__solver' : ['saga'],
            }
        ),
        
        "knn" : (
            KNeighborsClassifier(n_jobs=-1),
            {
                'model__n_neighbors': list(range(3, 14)),
                'model__p': [1, 2],
                'model__weights': ['uniform', 'distance'],
                'model__algorithm' : ['ball_tree', 'kd_tree', 'brute']
            }
        ),
        
        "SVM" : (
            SVC(probability=True, class_weight='balanced', random_state=42),
            {
                'model__C': [0.1, 1, 10],
                'model__kernel': ['linear', 'rbf', 'poly'],
                'model__gamma': ['scale', 'auto'],
                'model__degree': [2,3]
            }
        ),
        
        "random_forest" : (
            RandomForestClassifier(class_weight='balanced', random_state=42),
            {
                'model__n_estimators': [300, 600],
                'model__max_depth': [None, 8, 14],
                'model__min_samples_split': [2, 8],
                'model__class_weight': [None, "balanced_subsample"]
            }
        )
    }
    
    
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42, stratify=y)
    
    # Check the shape of x_train, x_test, y_train, y_test
    print('Check the shape of x_train, x_test, y_train, y_test')
    print(f'Shape of x_train: {x_train.shape}')
    print(f'Shape of y_train: {y_train.shape}')
    print(f'Shape of x_test: {x_test.shape}')
    print(f'Shape of y_test: {y_test.shape}')
    
    
    cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
    
    rows, best_models = [], {}
    
    for name, (estimator, grid) in models_and_grids.items():
        
        smote = SMOTE(
        sampling_strategy=0.5,   # minority becomes 0.3 the majority 
        k_neighbors=3         # reduce the number of neighbors
        )


        enn = EditedNearestNeighbours(
        n_neighbors=5,
        kind_sel="all"   # Less aggressive than 'all'
        )
        
        
        pipe = imbpipeline([
            
            ('prepro', preprocess),
            # ('smote', SMOTE(sampling_strategy=0.5, random_state=42)),
            ('smote_clean', SMOTEENN(smote=smote, enn=enn, random_state=42)),     # Generates data and then repeatedly deletes the corrupted data.
            ('model', estimator)
        ])
        
        grid_search = GridSearchCV(estimator=pipe, param_grid=grid, scoring='roc_auc', cv=cv, n_jobs=-1, verbose=2, refit=True)
        
        grid_search.fit(x_train, y_train)
        
        best_models[name] = grid_search.best_estimator_
        
        y_proba_train = grid_search.predict_proba(x_train)[:, 1]
        y_pred50_train = (y_proba_train >= 0.5).astype(int)
        
        
        y_proba = grid_search.predict_proba(x_test)[:, 1]
        y_pred50 = (y_proba >= 0.5).astype(int)
        
        rows.append({
            "Model": name,
            "BestParams": grid_search.best_params_,
            "CV_ROC_AUC": grid_search.best_score_,
            "Train_ROC_AUC": roc_auc_score(y_train, y_proba_train),
            "Train_Accuracy": accuracy_score(y_train, y_pred50_train),
            "Train_Precision": precision_score(y_train, y_pred50_train, zero_division=0),
            "Train_Recall": recall_score(y_train, y_pred50_train, zero_division=0),
            "Train_F1": f1_score(y_train, y_pred50_train),
            
            "Test_ROC_AUC": roc_auc_score(y_test, y_proba),
            "Test_Accuracy": accuracy_score(y_test, y_pred50),
            "Test_Precision": precision_score(y_test, y_pred50, zero_division=0),
            "Test_Recall": recall_score(y_test, y_pred50, zero_division=0),
            "Test_F1": f1_score(y_test, y_pred50)
        })
        
        
    results = pd.DataFrame(rows).sort_values('Test_ROC_AUC', ascending=False)
    
    best_row = results.iloc[results['Test_ROC_AUC'].values.argmax()]
    best_name = best_row['Model']
    best_model = best_models[best_name]
    
    # choose threshold on TRAIN only (no test leakage) 
    proba_oof = cross_val_predict(best_model, x_train, y_train, method='predict_proba')[:, 1]
    thresholds = np.linspace(0.2, 0.9, 20)
    # .argmax() returns the index of the first occurrence of the maximum value in an array.
    t_star = thresholds[np.argmax(f1_score(y_train, (proba_oof >= t).astype(int)) for t in thresholds)]
    
    # Evaluate on TEST with that fixed threshold
    y_proba = best_model.fit(x_train, y_train).predict_proba(x_test)[:, 1]
    y_pred_tuned = (y_proba >= t_star).astype(int)
    
    print(f"\nBest model: {best_name}")
    print(f"Best params: {best_row['BestParams']}")
    print(f"ROC AUC (Test): {best_row['Test_ROC_AUC']:.3f}")
    print(f"Chosen threshold for best F1 = {best_row['Test_F1']} in [{thresholds[0]:.1f}, {thresholds[-1]:.1f}]: {t_star:.2f}")
    
    # Check if there Overfitting or not
    print('\nCheck if there Overfitting or not')
    display(results)
    
    # reports & plots
    print("\nClassification Report (threshold=0.50):")
    print(classification_report(y_test, (y_proba >= 0.5).astype(int), digits=3))
    print("\nClassification Report (tuned threshold):")
    print(classification_report(y_test, y_pred_tuned, digits=3))
    
    ConfusionMatrixDisplay(confusion_matrix(y_test, y_pred50)).plot(values_format='d')
    plt.title(f"{best_name} — Confusion Matrix (thr={0.5:.2f})")
    plt.show()
    
    
    ConfusionMatrixDisplay(confusion_matrix(y_test, y_pred_tuned)).plot(values_format='d')
    plt.title(f"{best_name} — Confusion Matrix (thr={t_star:.2f})")
    plt.show()

    RocCurveDisplay.from_estimator(best_model, x_test, y_test)
    plt.title(f"{best_name} — ROC Curve (Test)")
    plt.show()

    PrecisionRecallDisplay.from_predictions(y_test, y_proba)
    plt.title(f"{best_name} — Precision-Recall (Test)")
    plt.show()
    
    
    def show_important(fitted_pipeline: Pipeline, top_k: int=20):
        # This attribute allows you to access any step parameter by user given name. The keys are step names, and the values are step parameters.
        # This feature is particularly useful for inspecting the pipeline's configuration and understanding how each step contributes to the overall model.
        prep = fitted_pipeline.named_steps['prepro']
        model = fitted_pipeline.named_steps['model']
        
        # names of features after OneHotEncoding
        cat_cols = [col for col in x.columns if x[col].dtype == 'object' or str(x[col].dtype).startswith('category')]
        num_names = [c for c in x.columns if c not in cat_cols]
        cat_names = list(
            # The `get_feature_names_out` method in scikit-learn is used to retrieve the names of the feature names that were used during the transformation process.
            prep.named_transformers_['cat'].named_steps['encoding_onehot'].get_feature_names_out(cat_cols)
        )
        
        all_feature_names = np.array(num_names + cat_names)
        
        if hasattr(model, 'feature_importances_'):
            vals, kind = model.feature_importances_, 'Feature Importances'
        
        elif hasattr(model, 'coef_'):
            vals, kind = np.abs(model.coef_).ravel(), '|Coefficients|'
        
        else:
            print("This estimator does not expose importances/coefficients.")
            return
        
        
        feature_order = np.argsort(vals)[::-1][:top_k]
        important_df = pd.DataFrame({'Feature' : all_feature_names[feature_order], 'Importance' : vals[feature_order]})
        display(important_df)
        
        plt.figure(figsize=(8, 6))
        plt.barh(important_df['Feature'][::-1], important_df['Importance'][::-1])
        # plt.title()
        plt.tight_layout()
        plt.show()
        
    show_important(best_model, top_k=20)
    
    return results, best_model
```

## 8. Executive Summary & Business Recommendations 

### 8.1 Model Performance & Selection
After rigorous testing of multiple algorithms and handling the highly imbalanced nature of our attrition data, the **Linear Support Vector Machine (SVM) combined with SMOTEENN** emerged as the most practical model for our HR department. 

Instead of focusing solely on overall accuracy, we prioritized **Recall**, as the primary business goal is to proactively identify employees at risk of leaving before they actually resign. 
* **Key Metric:** At the default decision threshold (0.50), the model successfully identifies **~81%** of employees who are likely to leave (Recall: 0.809).
* **Trade-off:** While this proactive approach yields some false alarms (Precision: 0.352), the cost of conducting "stay interviews" with a few stable employees is far lower than the cost of suddenly losing top talent.

### 8.2 Key Drivers of Attrition (What makes them leave?)
By extracting the feature importances (coefficients) from our Linear SVM, we uncovered the hidden patterns driving employee turnover. The top risk factors are:

1. **Job Role (Laboratory Technician):** This specific role has the highest correlation with attrition, indicating potential issues with compensation, career growth, or daily work conditions in this department.
2. **OverTime (The Burnout Factor):** Employees working consistent overtime (`OverTime_Yes`) are highly likely to burn out and leave. Interestingly, employees strictly working no overtime (`OverTime_No`) also show a tendency to leave, possibly indicating disengagement or lack of challenging opportunities.
3. **Business Travel:** Extremes in travel policies push employees away. Both frequent travelers (`Travel_Frequently`) and non-travelers (`Non-Travel`) show higher attrition rates compared to those who travel rarely.
4. **Role Stagnation (`YearsInCurrentRole`):** The longer an employee stays in the exact same role without a change or promotion, the higher the risk of them seeking opportunities elsewhere.

### 8.3 Strategic Recommendations for HR
Based on the data, we recommend the following proactive retention strategies:
* **Targeted Review for Lab Techs:** Conduct immediate focus groups or compensation benchmarks specifically for the Laboratory Technician role to identify and resolve their unique pain points.
* **OverTime Management:** Implement stricter monitoring of OverTime hours to prevent burnout. Ensure workload is distributed evenly.
* **Career Mobility Programs:** Address "Role Stagnation" by encouraging internal mobility, lateral moves, or cross-training for employees who have been in the same role for over 3-4 years.
* **Proactive "Stay Interviews":** Utilize the model's predictions monthly to generate a "high-risk" list. HR business partners should conduct casual check-ins with these individuals to address concerns before a resignation letter is drafted.
